Code to filter out short loci based on optimal allele length threshold before running SISTR

```python
%pylab inline
import sys
sys.path.append("/storage/BonnieH/selection_project/helper_functions")
from ABC_functions import *
```

```python
def FilterShortLoci(inFile, outFile, total_thresh=0):
    
    allele_freqs_file = open(inFile, 'r')
    results = open(outFile, 'w')
    
    num_lines_before_filt = 0
    num_lines_after_filt = 0
    
    print('Loci where freqs were not called')
    
    for line in allele_freqs_file:
        num_lines_before_filt = num_lines_before_filt + 1
        
        # Get information from line
        info = line.strip().split('\t')
        chrom = info[0]
        start = int(info[1])
        end = int(info[2])
        freqs = info[3]
        total = int(info[4])
        per = int(info[5])
        if freqs == '.':
            print(line)
        
        #  Only continue if freqs != '.'
        if freqs != '.' and total > total_thresh:  # total_thresh is 5214 for SSC data
            
            # Get info about reference allele (length in base pairs, repeat units)
            optimal_ru = Process_Freqs(freqs, per, end, start, False, True)
            
            if (per == 2 and optimal_ru >= 11) \
            or (per == 3 and optimal_ru >= 5) \
            or (per == 4 and optimal_ru >= 7): # or (per == 1 and optimal_ru >= 12):
                results.write(line)
                num_lines_after_filt = num_lines_after_filt + 1
        
    results.close()        
    allele_freqs_file.close()
    print("Done")
    print('Num loci before filtering: ' + str(num_lines_before_filt))
    print('Num loci after filtering: ' + str(num_lines_after_filt))
```

```python
def main():
    inFile = '/storage/BonnieH/selection_project/1000genomes/euro_allele_freqs/allele_freqs_annot_pers234.txt'
    outFile = '/storage/BonnieH/selection_project/1000genomes/euro_allele_freqs/allele_freqs_filt_w_total_thresh.txt'
    total_thresh = int(534*0.8)
    FilterShortLoci(inFile, outFile, total_thresh)
```

```python
%%time
if __name__ == '__main__':
    main()
```
