```python
%pylab inline
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
PLOTDIR = '/storage/BonnieH/selection_project/figures/figures/'
MAX_LOG_MU = -3
MIN_LOG_MU = -8
```

```python
def getMuList(x, mu, L, ru, per):
    mu_list = []
    for elem in x:
        
        log_mu = np.log10(mu)+L*(elem/per-ru)

        if log_mu < MIN_LOG_MU:
            log_mu = MIN_LOG_MU

        if log_mu > MAX_LOG_MU:
            log_mu = MAX_LOG_MU

        mu_list.append(log_mu)
    return mu_list
```

```python
# Plot homopolymers

x = list(range(9, 49)) # X axis - optimal allele length

# mu, L, repeat unit corresponding to mu, period

mu_list_1 = getMuList(x, 10**-3.6, 0.04, 13, 1)
mu_list_2 = getMuList(x, 10**-3.8, 0.04, 13, 1)
mu_list_3 = getMuList(x, 10**-4, 0.04, 13, 1)
mu_list_4 = getMuList(x, 10**-4.2, 0.04, 13, 1)
mu_list_5 = getMuList(x, 10**-4.4, 0.04, 13, 1)
mu_list_6 = getMuList(x, 10**-4.6, 0.08, 13, 1)
mu_list_7 = getMuList(x, 10**-4.8, 0.08, 13, 1)

plt.figure(1)
plt.plot(x,mu_list_1,label='Setting a', color='red')
plt.plot(x,mu_list_2,label='Setting b', color='green')
plt.plot(x,mu_list_3,label='Setting c', color='blue')
plt.plot(x,mu_list_4,label='Setting d', color='purple')
plt.plot(x,mu_list_5,label='Setting e', color='black')
plt.plot(x,mu_list_6,label='Setting f', color='orange')
plt.plot(x,mu_list_7,label='Setting g', color='c')
plt.xlabel('Optimal allele length (bp)', size=15)
plt.ylabel('Log 10 mutation rate', size=15)
plt.title('Hompolymer mutation rates used', size=15)
plt.xticks(fontsize=15) 
plt.yticks(fontsize=15)
plt.legend()
#plt.savefig(PLOTDIR + 'actual_mut_rates.pdf', bbox_inches='tight')
```
