# Supplemental Figure: 
### Plot distribution of TMRCA values for European demographic model (Tennessean et al (2012))

```python
%pylab inline
import numpy as np
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
from matplotlib import pyplot as plt

# Get list of TMRCA values
TMRCAFile = '/storage/BonnieH/selection_project/annotations/TMRCA.txt'
TMRCA_file = open(TMRCAFile, 'r')
TMRCA_list = []
for line in TMRCA_file:
        
    TMRCA = line.strip()
    TMRCA = float(TMRCA)
    TMRCA = int(TMRCA)
    
    TMRCA_list.append(TMRCA)
    
print('Mean TMRCA value: ' + str(np.mean(TMRCA_list)))
print('Max TMRCA value: ' + str(max(TMRCA_list)))
print('Min TMRCA value: ' + str(min(TMRCA_list)))
plt.hist(TMRCA_list, bins=range(0,120000,4000), color="black", edgecolor="white", weights=np.ones(len(TMRCA_list)) / len(TMRCA_list))
plt.title('Distribution of TMRCA values')
plt.xlabel('TMRCA')
plt.ylabel('Fraction of values')
plt.axvline(x=55920, linestyle = '--', label = 'TMRCA value used under constant\nnumber of generations simplification')
plt.legend()
plt.savefig('TMRCA_distribution.pdf', bbox_inches='tight')
```
