This notebook contains code to plot mutation rates for SISTR2

```python
%pylab inline
import os
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
PLOTDIR = 'pdfs-round1'
PREFIX = "Figure2-"
MAX_LOG_MU = -3
MIN_LOG_MU = -8
```

```python
def getMuList(x, mu, L, ru, per):
    mu_list = []
    for elem in x:
        
        log_mu = np.log10(mu)+L*(elem/per-ru)

        if log_mu < MIN_LOG_MU:
            log_mu = MIN_LOG_MU

        if log_mu > MAX_LOG_MU:
            log_mu = MAX_LOG_MU

        mu_list.append(log_mu)
    return mu_list
```

```python
# Plot dinucleotides

x = list(range(9, 49)) # X axis - optimal allele length

mu_list_0 = getMuList(x, 10**-3.5, 0.076, 6, 2)
mu_list_1 = getMuList(x, 10**-3.45, 0.096, 10, 2)
mu_list_2 = getMuList(x, 10**-4.2, 0.12, 6, 2)
mu_list_3 = getMuList(x, 10**-4.6, 0.135, 6, 2)
mu_list_4 = getMuList(x, 10**-5, 0.15, 6, 2)
mu_list_5 = getMuList(x, 10**-5.4, 0.165, 6, 2)

plt.figure(1)
plt.plot(x,mu_list_0,label='Setting a', color='red') # Used to start with z and end with f
plt.plot(x,mu_list_1,label='Setting b', color='green')
plt.plot(x,mu_list_2,label='Setting c', color='blue')
plt.plot(x,mu_list_3,label='Setting d', color='purple')
plt.plot(x,mu_list_4,label='Setting e', color='brown')
plt.plot(x,mu_list_5,label='Setting f', color='black')
plt.xlabel('Optimal allele length (bp)', size=15)
plt.ylabel('Log 10 mutation rate', size=15)
plt.title('Dinucleotide mutation rates used', size=15)
plt.xticks(fontsize=15) 
plt.yticks(fontsize=15)
plt.legend()
plt.savefig(os.path.join(PLOTDIR, PREFIX + 'dinucleotide_mut_rates.pdf'), bbox_inches='tight')
```
