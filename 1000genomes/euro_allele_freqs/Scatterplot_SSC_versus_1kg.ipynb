```python
%pylab inline
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from heapq import nlargest
from operator import itemgetter 
```

```python
inFile = '/storage/BonnieH/selection_project/1000genomes/euro_allele_freqs/1kg_scores_merged.txt'
inFileSSC = '/storage/BonnieH/selection_project/1000genomes/euro_allele_freqs/SSC_scores_merged.txt'

x = []
y = []

allele_freqs_file = open(inFile, 'r')
header = allele_freqs_file.readline().strip().split('\t')
    
for line in allele_freqs_file:
        
    # Get information from line
    info = line.strip().split('\t')
    score = info[3]
    if score == 'N/A':
        x.append(-1)
    elif score == '0':
        x.append(10**-5.1)
    else:
        x.append(float(score))
allele_freqs_file.close()

allele_freqs_file = open(inFileSSC, 'r')
header = allele_freqs_file.readline().strip().split('\t')
    
for line in allele_freqs_file:
        
    # Get information from line
    info = line.strip().split('\t')
    score = info[3]
    if score == 'N/A':
        y.append(-1)
    elif score == '0':
        y.append(10**-5.1)
    else:
        y.append(float(score))
allele_freqs_file.close()

plt.scatter(x,y,s=0.2)
plt.yscale('log')
plt.xscale('log')
plt.xlim(10**-5.2,0.3)
plt.ylim(10**-5.2,0.3)
plt.plot([10**-5.2,0.3],[10**-5.2,0.3],c='black')
plt.xlabel('1kg s value')
plt.ylabel('SSC s value')
plt.title('1kg versus SSC Per-locus SISTR scores')
```
